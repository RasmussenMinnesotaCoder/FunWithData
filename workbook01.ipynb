## Does having an attractive beauty factor increase the chance of getting a positive rating by students as it relates to teaching?

```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
beauty = sm.datasets.get_rdataset("TeachingRatings", "AER")

beauty_scores = beauty.data['beauty']
eval_scores = beauty.data['eval']
plt.scatter(eval_scores, beauty_scores)
plt.xlabel('Evaluation')
plt.ylabel('Beauty')

X = sm.add_constant(eval_scores)
results = sm.OLS(beauty_scores, X).fit()
results.summary()

```
